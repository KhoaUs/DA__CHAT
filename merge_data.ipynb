```python
import pandas as pd
import numpy as np
```

## **1. Shopee & Lazada**

```python
# Load Data
shopee_data = pd.read_csv('Data/shopee data.csv')
lazada_data = pd.read_csv('Data/LAZADA_MOCK_4000.csv')
```

```python
shopee_data.head()
```

```python
shopee_data.info()
```

```python
lazada_data.head()
```

```python
lazada_data.info()
```

```python
# Merge Data
merged_data = pd.concat([lazada_data, shopee_data], ignore_index=True)
merged_data.info() 
```

```python
# Chuẩn hóa tên cột của Shopee
shopee_data = shopee_data.rename(columns={
    'product name': 'product_name',
    'unit sold': 'sold',
    'review count': 'review_count',
    'seller name': 'seller_name',
    'category': 'categories'
})

# Thêm cột 'sku' vào Shopee để đồng bộ với Lazada
shopee_data['sku'] = None

# Thêm cột 'platform' để phân biệt nguồn dữ liệu
lazada_data['platform'] = 'Lazada'
shopee_data['platform'] = 'Shopee'

# Danh sách cột chung 
common_columns = ['platform', 'categories', 'sku', 'product_name', 'sold', 
                  'review_count', 'rating', 'price', 'currency', 
                  'seller_name', 'brand', 'url']

# Chỉ giữ các cột chung ở cả hai DataFrame
lazada_data = lazada_data[common_columns]
shopee_data = shopee_data[common_columns]

# Gộp hai bảng lại
merged_data = pd.concat([lazada_data, shopee_data], ignore_index=True)

print(merged_data.info())
```

```python
merged_data.head()
```

## **2. Merge Data - Tiki**

```python
tiki_data = pd.read_csv('Data/tiki__products.csv')
tiki_data.head()
```

```python
tiki_data.info()
```

```python
# Thêm cột platform
tiki_data['platform'] = 'Tiki'

# Đảm bảo thứ tự cột giống merged_data
tiki_data = tiki_data[['platform', 'categories', 'sku', 'product_name', 'sold',
                       'review_count', 'rating', 'price', 'currency',
                       'seller_name', 'brand', 'url']]

# Gộp dữ liệu lại
merged_data_2 = pd.concat([merged_data, tiki_data], ignore_index=True)
```

```python
merged_data_2.info()
```

## **3. Merge Data - TikTokShop**

```python
tiktokshop_data = pd.read_csv('Data/tts_cleaned.csv')
tiktokshop_data.head()
```

```python
tiktokshop_data.info()
```

```python
# Chuẩn hóa dữ liệu TikTok Shop
tiktokshop_data['categories'] = (
    tiktokshop_data['category'].fillna('') + 
    ' > ' + 
    tiktokshop_data['sub_category'].fillna('')
)

# Thêm các cột còn thiếu
tiktokshop_data['currency'] = 'VND'
tiktokshop_data['platform'] = 'TikTok Shop'

# Đổi tên cột để khớp với cấu trúc chung
tiktokshop_data = tiktokshop_data.rename(columns={
    'title': 'product_name',
    'sold_count': 'sold',
    'review_count': 'review_count',
    'score': 'rating',
    'sale_price': 'price',
    'sku_id': 'sku',
    'shop_name': 'seller_name',
    'canonical_url': 'url'
})

# Thêm cột 'brand' có giá trị giống 'seller_name'
tiktokshop_data['brand'] = tiktokshop_data['seller_name']

# Chọn đúng thứ tự cột giống final_merged
tiktokshop_data = tiktokshop_data[['platform', 'categories', 'sku', 'product_name', 'sold',
                                   'review_count', 'rating', 'price', 'currency',
                                   'seller_name', 'brand', 'url']]

# Gộp với dữ liệu trước (Lazada + Shopee + Tiki)
final_merged_all = pd.concat([merged_data_2, tiktokshop_data], ignore_index=True)
```

```python
final_merged_all.info()
```

```python
final_merged_all['sku'] = final_merged_all['sku'].astype(str)
final_merged_all['price'] = final_merged_all['price'].astype(float)
```

```python
# Save to CSV
final_merged_all.to_csv('Data/final_merged_all.csv', index=False)
```
