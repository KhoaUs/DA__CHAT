Kiểm tra cột `categories` theo `platform`

```python
def check_categories(data, platform):
    print(f'\n Number of categories in {platform}:', len(data[data['platform'] == platform]['categories'].unique()))
    categories = data[data['platform'] == platform]['categories'].unique()
    print('\nCategories for', platform, ':', categories)
    
check_categories(raw_df, 'Lazada')
check_categories(raw_df, 'Shopee')
check_categories(raw_df, 'Tiki')
check_categories(raw_df, 'TikTok Shop')
```

Chuẩn hóa cột `categories`

```python
df = raw_df.copy()

# Lọc dữ liệu TikTok Shop
tiktok_df = df[df["platform"] == "TikTok Shop"].copy()

# --- Hàm chuẩn hóa danh mục TikTok Shop ---
def clean_tiktok_category(cat):
    if pd.isna(cat):
        return None
    # Lấy nhánh cuối cùng sau dấu ">"
    last_part = cat.split(">")[-1].strip()
    # Thay dấu "-" bằng khoảng trắng
    last_part = last_part.replace("-", " ")
    # Loại bỏ ký tự đặc biệt (nếu có)
    last_part = re.sub(r"[^A-Za-z0-9 ]", "", last_part)
    # Viết hoa chữ cái đầu mỗi từ
    return last_part.title().strip()

# Áp dụng cho cột categories
tiktok_df["categories"] = tiktok_df["categories"].apply(clean_tiktok_category)
```

```python
df[df["platform"] == "TikTok Shop"] = tiktok_df
check_categories(df, 'TikTok Shop')
```
